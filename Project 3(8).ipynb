# 1. Introduction

Financial institutions process millions of financial transactions every day, making manual monitoring for suspicious activity increasingly difficult. Anti-Money Laundering (AML) systems are designed to identify transactions that may indicate money laundering, terrorist financing, or other financial crimes. Traditionally, financial institutions rely on rule-based transaction monitoring systems, where predefined rules generate alerts for further investigation.

Although rule-based systems are effective for detecting known suspicious patterns, they often produce a large number of false positives and may fail to detect new or evolving money laundering techniques. As financial transactions continue to increase in volume and complexity, machine learning has emerged as a promising approach for improving transaction monitoring by learning patterns directly from historical data.

This project investigates the use of supervised machine learning algorithms to classify financial transactions as suspicious or non-suspicious using the SAML-D synthetic transaction monitoring dataset. Several classification models are trained and compared to determine which algorithm performs best for AML transaction monitoring. The project also demonstrates the complete machine learning workflow, including data preprocessing, feature engineering, dimensionality reduction, model training and performance evaluation.

## 6.8 Dimensionality Reduction (PCA)

Principal Component Analysis (PCA) is applied to examine whether the dimensionality of the feature space can be reduced while retaining most of the variance in the data.

PCA is applied only to the standardized numerical features to ensure that all variables contribute proportionally to the transformation.

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train)

explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(cumulative_variance)+1),
    cumulative_variance,
    marker="o"
)

plt.axhline(y=0.95, color="r", linestyle="--", label="95% variance")
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.legend()
plt.show()
```

**Observation**

The cumulative explained variance curve shows that approximately **8 out of the 11 available components** are required to capture 95% of the total variance in the dataset. This indicates that the features carry relatively distributed, non-redundant information rather than being dominated by a small number of components.

This result is consistent with the earlier correlation analysis (Section 5.12), where numerical features showed weak linear correlations with one another. Since PCA works best when features are correlated, the relatively high number of components needed to reach 95% variance confirms that most features contribute unique, non-overlapping information, and the dataset does not compress well into a lower-dimensional space.

As a result, dimensionality reduction offers limited benefit for this dataset, and the original feature set is retained for model training rather than the PCA-transformed components.

```python
pca_2d = PCA(n_components=2)
X_train_pca = pca_2d.fit_transform(X_train)

plt.figure(figsize=(8,6))

plt.scatter(
    X_train_pca[:,0],
    X_train_pca[:,1],
    c=y_train,
    cmap="coolwarm",
    alpha=0.3,
    s=5
)

plt.title("PCA - First Two Components by Class")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Is_laundering")
plt.show()
```

### Observation

The PCA scatter plot shows that the first two principal components do not clearly separate legitimate and laundering transactions. A substantial overlap exists between the two classes, indicating that the original features cannot be effectively represented in only two dimensions.

This result suggests that PCA does not improve class separability for this dataset. Although dimensionality reduction is useful for visualization, important discriminatory information is distributed across multiple features rather than concentrated in the first two principal components.

Therefore, the original feature set is retained for model development instead of replacing it with the PCA-transformed components.

## 7.4 Support Vector Machine (LinearSVC)

Support Vector Machines aim to find the optimal hyperplane that separates the two classes with maximum margin.

LinearSVC is used instead of the standard SVC implementation because it scales significantly better to large datasets while still providing strong performance on linearly separable or near-linearly separable data.

Class weighting is applied to help the model account for the imbalance between legitimate and laundering transactions.

```python
from sklearn.svm import LinearSVC

svm_model = LinearSVC(
    class_weight="balanced",
    random_state=42,
    max_iter=5000,
    dual=False
)

svm_model.fit(X_train, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)

# LinearSVC has no predict_proba by default, use decision_function for ROC/AUC
y_score_svm = svm_model.decision_function(X_test)
```

### Model Evaluation

```python
print("Accuracy :", round(accuracy_score(y_test, y_pred_svm),4))
print("Precision:", round(precision_score(y_test, y_pred_svm),4))
print("Recall   :", round(recall_score(y_test, y_pred_svm),4))
print("F1-score :", round(f1_score(y_test, y_pred_svm),4))
print("ROC-AUC  :", round(roc_auc_score(y_test, y_score_svm),4))
```

```python
cm_svm = confusion_matrix(y_test, y_pred_svm)

ConfusionMatrixDisplay(cm_svm).plot()
plt.title("LinearSVC Confusion Matrix")
plt.show()
```

```python
print(classification_report(y_test, y_pred_svm))
```

### Observation

Unlike the previous models, class weighting causes LinearSVC to prioritize detecting the minority class rather than maximizing overall accuracy. This results in a considerably lower accuracy (74.7%) compared to Logistic Regression, Decision Tree and Random Forest.

However, Recall improves substantially to 45.52%, the highest among all models evaluated so far, meaning LinearSVC identifies a much larger proportion of actual laundering transactions. This comes at the cost of very low Precision (9.05%), as the model also produces a large number of false positives (9,038 legitimate transactions misclassified as suspicious).

The resulting F1-score (0.1509) and ROC-AUC (0.6717) are modest and comparable to Logistic Regression, indicating that while class weighting improves the model's ability to flag suspicious activity, it does not necessarily lead to a better overall balance between Precision and Recall.

These results illustrate an important trade-off in AML detection: models tuned to maximize Recall (catching more laundering cases) tend to generate more false alarms, which increases the investigative workload for compliance analysts.

# 8.1 ROC Curve Comparison

Receiver Operating Characteristic (ROC) curves provide a graphical comparison of the discriminative performance of the evaluated machine learning models.

The ROC curve illustrates the trade-off between the True Positive Rate (Recall) and the False Positive Rate across different classification thresholds. A model whose curve is closer to the upper-left corner generally demonstrates better classification performance.

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# Logistic Regression
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
plt.plot(fpr_lr, tpr_lr,
         label=f"Logistic Regression (AUC = {roc_auc_score(y_test, y_prob_lr):.3f})")

# Decision Tree
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
plt.plot(fpr_dt, tpr_dt,
         label=f"Decision Tree (AUC = {roc_auc_score(y_test, y_prob_dt):.3f})")

# Random Forest
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
plt.plot(fpr_rf, tpr_rf,
         label=f"Random Forest (AUC = {roc_auc_score(y_test, y_prob_rf):.3f})")

plt.plot([0,1],[0,1],"k--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")

plt.legend()

plt.show()
```

### Observation

The ROC curves confirm that Random Forest provides the strongest overall discriminative performance, achieving the largest area under the curve (ROC-AUC). Logistic Regression performs moderately well but struggles to identify the minority class, while Decision Tree demonstrates lower overall discrimination.

These findings are consistent with the quantitative evaluation presented in the previous section and support the selection of Random Forest as the best-performing model.

# 8.2 Precision–Recall Curve Comparison

Because money laundering detection involves a relatively rare positive class, Precision–Recall (PR) curves provide a more informative evaluation than Accuracy alone.

The PR curve illustrates the trade-off between Precision and Recall across different classification thresholds and is particularly useful for imbalanced classification problems such as AML transaction monitoring.

```python
from sklearn.metrics import precision_recall_curve

plt.figure(figsize=(8,6))

# Logistic Regression
precision_lr, recall_lr, _ = precision_recall_curve(y_test, y_prob_lr)
plt.plot(recall_lr, precision_lr, label="Logistic Regression")

# Decision Tree
precision_dt, recall_dt, _ = precision_recall_curve(y_test, y_prob_dt)
plt.plot(recall_dt, precision_dt, label="Decision Tree")

# Random Forest
precision_rf, recall_rf, _ = precision_recall_curve(y_test, y_prob_rf)
plt.plot(recall_rf, precision_rf, label="Random Forest")

plt.xlabel("Recall")
plt.ylabel("Precision")

plt.title("Precision–Recall Curve Comparison")

plt.legend()

plt.show()
```

### Observation

The Precision–Recall curves further highlight the differences between the evaluated models. Random Forest maintains the most balanced relationship between Precision and Recall, while Decision Tree achieves higher Recall at the expense of Precision. Logistic Regression detects only a small proportion of laundering transactions.

The Precision–Recall analysis confirms that Random Forest provides the most suitable balance between identifying suspicious transactions and minimizing false positive alerts.
